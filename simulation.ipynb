```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_scores = pd.read_csv("all_scores.csv")
mean = np.mean(all_scores['Score'])
std = np.std(all_scores['Score'])

sample_normal = []

for i in np.arange(95):
    normal = np.random.normal(mean, std)
    sample_normal.append(normal)

sample_lognormal = []

for i in np.arange(95):
    lognormal = np.random.normal(mean, std)
    sample_lognormal.append(lognormal)
```

```python
plt.hist(all_scores['Score'], alpha = 0.5, label='Actual Scores', color = 'blue') 
plt.hist(sample_normal, alpha = 0.5, label='Simulated Normal Scores', color = 'red') 
plt.hist(sample_lognormal, alpha = 0.5, label='Simulated Log-Normal Scores', color = 'yellow') 
  
plt.legend(loc='upper left') 
plt.title('Overlapping') 
plt.show()
```
